```python
## How to download data from google drive using gdown
import gdown
```

```python
url = "https://drive.google.com/file/d/16PZpADG4Pl_SBr2E3DEcvXsLQ5DSUtDP/view?usp=sharing"
```

```python
file_id = url.split("/")[-2]
file_id
```

```python
prefix = "https://drive.google.com/uc?export=download&id="
gdown.download(prefix + file_id, "kidney_data.zip", quiet=False)
```
